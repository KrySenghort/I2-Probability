### <font color = yellow>Conditional Probability <br></font>
<font color = green|>If A and B are two dependent events to each other, then probability is given by :
$$ P(B/A) = \frac{P(A \cap B)}{P(A)} $$
Where $ P(B/A) $ is the probability of B given that A has occurred <br>
      $ P(A \cap B) $ is the probability of A and B both occurring

For example : 
- I give my students two tests. 60% of my students passed both tests, but the first test was easier that 80% of students passed that one. What percentage of students who passed the first test also passed the second ? 
- A = passing the first test, B = passing second test
- So we are asking for P(B/A) 
- $$ P(B/A) = \frac{P(A \cap B)}{P(A)} = \frac{0.6}{0.8} = 0.75 $$
- Therefore, 75% of students who passed the first test passed the second.

- A: 1 on the first throw, 2 on the second throw, and 3 on the third throw? 
- B: 1, 2, and 3 in any order? 
- C: Three of 6 ?

````python
# A: 1 on the first throw, 2 on the second throw, and 3 on the third throw? 
success = 0
tries = 0
for first in range (0,6) :
  for second in range (0,6) :
      for third in range (0,6):
          if first == 1 and second == 2 and third == 3:
              success+=1
          tries+=1
print (float(success)/tries)
````

````python
# B: 1, 2, and 3 in any order?
success = 0
tries = 0
for first in range(0,6):
    for second in range(0,6):
        for third in range(0,6):
            if first == 1 :
                if second == 2 :
                    if third == 3:
                        success+=1
                elif second == 3:
                    if third == 2 :
                        success+=1
            elif first == 2 :
                if second == 1 :
                    if third == 3:
                        success+=1
                elif second == 3:
                    if third == 1 :
                        success+=1
            elif first == 3 :
                if second == 1 :
                    if third == 2:
                        success+=1
                elif second == 2:
                    if third == 1 :
                        success+=1                               
            tries+=1
print (float(success)/tries)
#Ans : 0.027777777777777776 i.e. 6/6^3 = 1/6^2
````

````python
#  C: Three 6 ? 
success = 0
tries = 0
for first in range(1,7):
    for second in range(1,7):
        for third in range(1,7):
            if first == 6 and second == 6 :
                if third == 6 :
                    success+=1
            tries+=1
print ('Success=',success)
print ('tries=',tries)
print ('Probability=',float(success)/tries)
#Ans = 0.004629629629629629 i.e. 1/6^3 
````

### <font color = yellow>III.) Discrete Random Variable & Probability Distribution

A discrete random variable is a variable that can take on only a finite or countable number of values. For example, the number of heads you get when you flip a coin is a discrete random variable. It can only take on the values 0 or 1.

The probability distribution of a discrete random variable is a list of all the possible values that the variable can take on, along with the probability of each value occurring. For example, the probability distribution of the number of heads you get when you flip a coin is:

```
Value | Probability
------- | --------
0     | 0.5
1     | 0.5
```

The sum of all the probabilities in a probability distribution must be equal to 1. This is because the probability of any event occurring must be between 0 and 1, inclusive.

Discrete random variables and probability distributions are used in many areas of mathematics and statistics. They are used to model a wide variety of phenomena, such as the number of cars that pass through an intersection in a given hour, the number of students who get a certain grade on a test, and the number of times a coin lands on heads in a series of flips.

Here are some examples of discrete random variables:

* The number of heads you get when you flip a coin
* The number of times you roll a 6 on a standard 6-sided die
* The number of people in a room who have brown hair
* The number of cars that pass through an intersection in a given hour
* The number of students who get an A on a test

These are just a few examples of the many possible discrete random variables. Discrete random variables are a common occurrence in the real world, and they can be used to model a wide variety of phenomena.

The probability distribution of the number of heads you get when you flip a coin 10 times can be calculated using the following steps:

1. Define a function that flips a coin 10 times and returns the number of heads.
2. Use a for loop to iterate over all possible values of the number of heads, from 0 to 10.
3. In each iteration, call the function that flips the coin 10 times and store the result in a variable.
4. Calculate the probability of getting the observed number of heads by dividing the number of times the function returned the observed number of heads by the total number of iterations.
5. Plot the probability of getting each number of heads as a bar graph.

The following Python code implements these steps:

````python
import random
import matplotlib.pyplot as plt
def flip_coin(n):
  """Flips a coin n times and returns the number of heads."""
  heads = 0
  for i in range(n):
    if random.random() < 0.5:
      heads += 1
  return heads

probabilities = []
for i in range(11):
  probabilities.append(flip_coin(10) == i)

plt.bar(range(11), probabilities)
plt.xlabel("Number of heads")
plt.ylabel("Probability")
plt.show()
````

<font color = green|>Exercise 03

Find the probability distributions of the number of people in a room who have brown hair.

````python
import random

def binomial_distribution(n, p):
  """Calculates the probability distribution of the number of successes in a sequence of n independent experiments, each of which has a constant probability of success of p.

  Args:
    n: The number of experiments.
    p: The probability of success.

  Returns:
    A list of probabilities, one for each possible number of successes.
  """
  probabilities = []
  for i in range(n + 1):
    probabilities.append(choose(n, i) * p**i * (1 - p)**(n - i))
  return probabilities

def choose(n, k):
  """Calculates the binomial coefficient n choose k.

  Args:
    n: The total number of experiments.
    k: The number of successes.

  Returns:
    The binomial coefficient n choose k.
  """
  if k > n:
    return 0
  numerator = 1
  denominator = 1
  for i in range(k):
    numerator *= n - i
    denominator *= i + 1
  return numerator // denominator

probabilities = binomial_distribution(10, 0.5)

print(probabilities)

plt.hist(probabilities, bins=6)
plt.ylabel("Number on die")
plt.xlabel("Probability")
plt.show()
````

Find the probability of he number of cars that pass through an intersection in a given hour.

$$ pmf = \frac{(rate * time)^{k}*e^{-rate * time}}{k!} $$

````python
import math

rate = 2 # cars per hour
time = 1 # hour 
k = 4 # number of cars passing through in the given hour

# Calculate Poisson probability mass function
pmf = ((rate * time)**k * np.exp(-rate * time)) / math.factorial(k)
print("The probability of exactly {} cars passing through the intersection in an hour is {:.2%}".format(k, pmf))
````

````python
from scipy.stats import poisson

# Set the rate of cars passing through the intersection per hour
rate = 200

# Simulate the number of cars passing through the intersection in one hour
num_cars = np.random.poisson(rate)

# Print the number of cars
print("Number of cars passing through the intersection in one hour:", num_cars)

# Generate a Poisson distribution for the number of cars passing through the intersection in one hour
x = np.arange(poisson.ppf(0.01, rate), poisson.ppf(0.99, rate))
poisson_dist = poisson.pmf(x, rate)

# Plot the Poisson distribution
plt.plot(x, poisson_dist, 'bo', ms=8, label='Poisson PMF')
plt.vlines(x, 0, poisson_dist, colors='b', lw=5, alpha=0.5)
plt.title("Poisson Distribution for Number of Cars Passing Through Intersection in an Hour")
plt.xlabel("Number of cars")
plt.ylabel("Probability mass")
plt.legend()
plt.show()
````

## <font color = yellow>VI.)Hypergeometric Distribution

The hypergeometric experiment is an experiment satisfying the
following properties:
1. The population or set to be sampled consists of N individuals, objects, or elements (a finite population).

2. Each individual can be characterized as a success (S) or a failure (F), and there are M successes in the population.

3. A sample of n individuals is selected without replacement in such a way that each subset of size n is equally likely to be chosen.

The hypergeometric random variable X associated with hypergeometric experiment is defined by
- X = the number of success in the sample.
- X ∼ Hp(n, M, N) denotes the hypergeometric rv of sample size n drawn from a population of size N consisting of M success

If X ∼ Hp(n, M, N), then the pmf of the hypergeometrix rv X is given by
<font color = green|>
$$ P_x(x) = \frac{C_M ^x.C_{N−M} ^{n-x}}{C_N ^ n} $$
</font> 
for all integers x satisfying  
<font color = green|>

$$ max(0, n − N + M) ≤ x ≤ min(n, M) $$

If X ∼ Hp(n, M, N), then <font color = green|>
$$ E(X) = n.\frac{M}{N} $$
$$ V(x) = (\frac{N-n}{N-1})n\frac{M}{N}(1-\frac{M}{N}) $$

## <font color = yellow>XV.) Exponential Distribution

The exponential distribution is a continuous probability distribution that describes the time between events in a Poisson process, where events occur at a constant average rate. It has only one parameter, denoted by λ (lambda), which represents the rate of occurrence of events.

The probability density function (PDF) of the exponential distribution is given by:

- $ f(x) = \frac{1}{λ}e^{-\frac{x}{\lambda}} $ if $ x>=0 $

- $ f(x) = 0$ for $x < 0.$

The cumulative distribution function (CDF) of the exponential distribution is given by:

- $ F(x) = 1 - e^{-\frac{x}{λ}} $ if $ x>=0 $

- $ F(x) = 0$ for $x < 0.$

Some properties of the exponential distribution are:

- The mean of the distribution is $ \frac{1}{λ}.$
- The variance of the distribution is $\frac{1}{λ^2}.$
- The memoryless property: The probability of an event occurring in the next time interval is independent of how much time has elapsed since the last event occurred.

The exponential distribution is widely used in reliability engineering, queueing theory, and other fields where the time between events is important.

One example of the Exponential Distribution in the real world is the time between radioactive decay events. Radioactive decay occurs randomly, and the time between two events follows an exponential distribution. For example, if you have a sample of radioactive material, the time it takes for half of the atoms to decay, known as the half-life, follows an exponential distribution.

Another example could be the time it takes for a person to receive their next email or phone call. Assuming that emails or calls arrive randomly, the time between two consecutive emails or calls can be modeled using an exponential distribution.

Additionally, the Exponential Distribution can be used in various other fields such as finance, reliability engineering, and healthcare to model the time between successive events, failures or occurrences.

## <font color = yellow>XVI.) Gamma Distribution

The gamma distribution is a continuous probability distribution with two parameters, shape and scale, that is commonly used in statistical analysis. It is often used to model the waiting time between events, such as the time between phone calls at a call center or the inter-arrival time of vehicles at a toll booth.

The probability density function of the gamma distribution is given by:
$$ f(x; \alpha, \beta) = \frac{x^{\alpha-1} e^{-x/\beta}}{{\beta^\alpha Γ(\alpha)}}, x>=0 $$

where x is a positive random variable
- $\alpha$ is the shape parameter
- $\beta$ is the scale parameter
- $Γ(\alpha)$ is the gamma function.

The mean and variance of the gamma distribution are given by:
$$ E(X) = \alpha\beta $$
$$ Var(X) = \alpha\beta^2 $$

The gamma distribution has many useful properties and applications, and it is frequently used in fields such as engineering, physics, finance, and healthcare.

They can also calculate the mean and variance of the returns, which can be used to optimize the risk-reward tradeoff of the portfolio.

## <font color = yellow>XVII.) Chi-squared Distribution

The chi-squared distribution is a continuous probability distribution that is widely used in statistics to model the sum of squared standard normal random variables. The chi-squared distribution has one parameter, often denoted by k, which represents the degrees of freedom.

The probability density function (PDF) of the chi-squared distribution is given by:

$$f(x; k) = \frac{1}{2^{k/2}Γ(k/2)}x^{(k/2)-1} e^{-x/2}$$

where $x >= 0, k > 0,$ and $Γ(k/2)$ is the gamma function of $k/2.$

The mean and variance of the chi-squared distribution are μ = k and σ^2 = 2k, respectively.

In Python, you can use the scipy.stats module to work with the chi-squared distribution. Here's an example of how to generate random samples from a chi-squared distribution with 5 degrees of freedom:

They calculate the expected proportions assuming that there is no difference between the brands, and compare them to the observed proportions from the survey data. If the calculated chi-squared value exceeds a certain threshold based on the degrees of freedom and level of significance, they can reject the null hypothesis that there is no difference between the brands and conclude that there is a statistically significant difference.

## <font color = yellow>XVIII.) Beta Distribution

The beta distribution is a continuous probability distribution defined on the interval [0, 1]. It is commonly used to model random variables that represent proportions or probabilities.

The probability density function of the beta distribution is given by:

$$ f(x; α, β) = \frac{x^{α-1}(1-x)^{β-1}}{B(α, β)} $$

where x is a value between 0 and 1, α > 0 and β > 0 are shape parameters, and B(α, β) is the beta function, which serves as a normalization constant.

- mean of the beta distribution is $ \frac{α}{α + β} $ 
- variance is $ \frac{αβ}{(α + β)^2(α + β + 1)} $ 

The skewness and kurtosis of the distribution depend on the values of the shape parameters.
The beta distribution has many applications in statistics and other fields. For example, it can be used to model the success rate of a binary outcome, such as the probability of a coin landing heads or the proportion of voters supporting a certain candidate. It can also be used in Bayesian analysis as a prior distribution for parameters that are restricted to the interval [0, 1].

## <font color = yellow>XX.) Weibull Distribution

The Weibull distribution is a continuous probability distribution that is commonly used to model lifetimes, failure times, and reliability in various fields. It was named after Swedish mathematician Waloddi Weibull, who first introduced it in 1951.

The probability density function (PDF) of the Weibull distribution is given by:
$$
f(x;\lambda,k) = \begin{cases}
\frac{k}{\lambda} (\frac{x}{\lambda})^{k-1} e^{-(\frac{x}{\lambda})^k} & x\geq 0 \\
0 & x < 0
\end{cases}
$$
where x is the random variable, λ (lambda) is the scale parameter, and k is the shape parameter. The scale parameter determines the location of the distribution, while the shape parameter controls the rate at which the failure or event occurs.

The cumulative distribution function (CDF) of the Weibull distribution is given by:
$$
F(x;\lambda,k) = 1 - e^{-(\frac{x}{\lambda})^k}
$$
The Weibull distribution can be used to model various types of phenomena such as the time until failure of a mechanical component, the time until a patient recovers from an illness, or the time until a product reaches the end of its useful life.

Sure, the Weibull distribution is commonly used in reliability engineering to model the failure rate of a product over time.
